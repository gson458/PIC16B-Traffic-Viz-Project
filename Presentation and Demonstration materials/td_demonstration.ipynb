```python
import traffic_data as td
import sqlite3
```

The import can take a while (up to a minute) because of the code running to retrieve county boundaries.

If we don't need to display those, we can remove the code near the top of traffic_data and the relevant `update_layer` parts of the `incident_scattermap` and `incident_heatmap` functions.

```python
conn = sqlite3.connect('test.db')
```

`td.initialize_db()` has to be run for any new databases so that the counties table is created - if its already been run for the database you don't need to run it again.

```python
td.initialize_db(conn)
```

To update the database, just call `update_county_incidents()`. Don't use `update_county_incidents_synchronous()`, that's the old and slower method.

```python
td.update_county_incidents(conn, "los angeles")
```

As long as the code runs without errors you can pretty much just ignore printed stuff like `County N/A` and `Except triggered`; those are just for edge cases like an incident not having an address in a county or the incidents table having to be created for the first time.

```python
data = td.get_county_incidents(conn, "los angeles")
td.incident_scattermap(data)
```

Be careful of how the county name is written; don't include the word 'county' when using it as input. Orange County, for example should just be 'Orange'.

```python
td.update_county_incidents(conn, "orange")
```
